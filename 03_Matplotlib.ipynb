### Matplotlib이란
  - 파이썬에서 가장 널리 사용되는 시각화 라이브러리
  - 시각화?
    수치데이터를 이해하기 쉽도록 시각적으로 표현하고 전달하는 과정
  - 선 그래프(plot), 산점도(scatter plot), 히스토그램(Histogram) 등의 데이터 분석 시각화에 많이 사용되는 MATLAB 형태의 그래프를 그릴 수 있다.

- Matplotlib 특징
  - 표준 플롯을 쉽게 그릴 수 있을 뿐 아니라 세부적 변경, 복잡한 플롯도 가능하다
  - numpy, pandas에서 사용되는 자료구조를 쉽게 시각화 가능

```python
import matplotlib
print(matplotlib.__version__)
```

```python
import matplotlib.pyplot as plt
# from matplotlib import pyplot as plt 랑 동일
```

- plot() 함수
  - 수치데이터 -> 선 그래프
  - plt.show() 함수 호출하여 그래프 화면에 표시
  - 그래프의 타이틀, 축 이름설정 가능

```python
x = [ i**2 for i in range(1,9)]
plt.plot(x)
plt.plot(x,color='r')
plt.plot(x,'or')
plt.plot(x,'-k')
plt.plot(x,'^b')
plt.plot(x,'--oy')

plt.show()
```

```python
import matplotlib.pyplot as plt
x = [i**2 for i in range(1,9)]
y = [i for i in range(1,9)]

plt.plot(x,y)
plt.xlabel('x Line')
plt.ylabel('y Line')
plt.title('x y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,6,0.1)
y1=np.sin(x)
y2=np.cos(x)

plt.plot(x,y1,label='Sin')
plt.plot(x,y2,label='Cos', linestyle='--')
plt.xlabel('X')
plt.ylabel('Y')
plt.title(' Sin&Cos')
plt.legend() #앞의 라벨 중 잘 구분할 수 있도록
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

a = np.arange(0,12,0.01)
b = np.sin(a)
c = np.cos(a)

plt.figure(figsize=(10,6))
plt.plot(a,b,'g',label='Sin')
plt.plot(a,c,color='blue',linestyle='dashed', label='Cos')
plt.grid()
plt.legend()
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Example of plt')
plt.show()
```

```python
a = np.arange(0,9)
b = [1,4,6,9,5,2,7,3,8]

plt.figure(figsize=(10,6))
plt.plot(a,b,
         color='red', linestyle='dashed',
         marker='o',markerfacecolor='blue',
         markersize=12)
# plt.xlim()
# plt.ylim([2.5,7.5])
plt.show()
```

```python
a = np.arange(0,10)
b = np.array([9,4,6,9,5,2,7,1,3,2])

plt.figure(figsize=(10,6))
plt.scatter(x=a,y=b,s=80,c=b,marker='>')
plt.colorbar()
plt.show()
```

```python
a = np.random.normal(loc=0, scale=1, size=1000)
b = np.random.normal(loc=5, scale=0.5, size=1000)
c = np.random.normal(loc=10, scale=2, size=1000)

plt.figure(figsize=(10,6))
plt.plot(a,label='a')
plt.plot(b,label='b')
plt.plot(c,label='c')
plt.legend()
plt.show()
```

```python
import statistics
# data = [3,2,1,4,3,1,5,2,12,1]
# data = [1,2,3,4,5,6,7,8,9]
data = [4,2,6,8,3,9,5,11]
mean_val = statistics.mean(data)
median_val = statistics.median(data)

print(f"mean value : {mean_val}")
print(f"median value : {median_val}")
```

```python
a = np.random.normal(loc=0, scale=1, size=1000)
b = np.random.normal(loc=5, scale=0.5, size=1000)
c = np.random.normal(loc=10, scale=2, size=1000)

plt.figure(figsize=(10,6))
plt.boxplot((a,b,c))
plt.legend()
plt.show()
```

```python
studentName = ['Jane','Alice','Tom', 'John']
score = [85,100,75,60]

plt.figure(figsize=(10,6))
plt.bar(studentName,score)
plt.show()

a=np.arange(len(studentName))
sc1 = plt.subplot()
fig,sc1 = plt.subplots()
sc1.bar(a,score,color='green')
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)

plt.plot(x,y, 'b-.', linewidth=3)
plt.xticks(np.arange(0,10,2))
plt.yticks(np.arange(-1.0,1.1))
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)

plt.plot(x,y, linewidth=2)
plt.xticks([0,5,10], labels=['Jan','Feb','Mar'])
plt.yticks([-1,0,1], labels=['min','ave','max'])
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)

plt.plot(x,y, linewidth=3)
plt.xticks([0,5,10],labels=['Jan','Feb','Mar'], rotation=90)
plt.yticks([-1,0,1], labels=['min','ave','max'])
plt.show()
```
