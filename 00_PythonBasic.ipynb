- List 정렬 : sort(), reverse(), sorted()
- List 복사 : h1[:] #복사
- List 확장 : extend()

```python
a = [65,12,98,43,35]
a.sort()
print(a)

b = [65,12,98,43,35]
b.reverse();
print(b)

c = [65,12,98,43,35]
c.sort(); c.reverse()
print(c)
#sort, reverse : 원본을 건드림


d = [65,12,98,43,35]
d.sort(reverse=True)
print(d)

e = [65,12,98,43,35]
f = sorted(e)
print(e,f)
```

```python
g1 =100; g2 = 100
g3 = g4 = g5 = 100
print(g1,g2,g3,g4,g5)
print(id(g1), id(g2), id(g3), id(g4), id(g5))
```

```python
h1 = [65,12,98,43,35]
h2 = [65,12,98,43,35]
h3 = h1

print(id(h1), id(h2), id(h3))

h1[1] = 120
print(h1[1],h3[1],h2[1])

print(h1[:])

h4 = h1[:] #copy
print(id(h1), id(h2), id(h3), id(h4))
```

- Tuple type : Data 중복 가능, 수정 불가

```python
a = (1,2,3)
print(a,type(a))
#a[1] = 20 #수정불가
```

```python
b = (10,)       ; print(type(b),b)
b = 10,         ; print(type(b),b)
b = 'A','B','C' ; print(type(b),b)
b = 'A',        ; print(type(b),b)
b = 10,20,30    ; print(type(b),b)
b = 10,20,        ; print(type(b),b)
```

```python
b = (10,20),     ; print(type(b),b)
b = [10, 20],   ; print(type(b),b)
b = (10), [20],  ; print(type(b),b)
```

```python
c1 = 10, ; c2 = 20,
print(type(c1), type(c2))
print(c1+c2) # 동일한 타입 연산 가능
```

```python
d1 = (10) ; d2 = (10,) ; d3 = (10), ; d4 = 10,

print(type(d1), type(d2), type(d3), type(d4))
print(d1, d2, d3, d4)
```

```python
e1 = [10],
e2 = (11,None,)
e3 = None,None
e4 = None, None
print(e1,e2,e3,e4)
```
