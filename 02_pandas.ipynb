### 그룹핑
- groupby() : 특정 칼럼의 값을 기준으로 그룹핑해서 함수를 실행한다.
- map : 단일컬럼에서 apply()랑 같은 기능


```python
df = pd.DataFrame({
'city': ['Busan', 'Busan', 'Busan', 'Busan',
'Seoul', 'Seoul', 'Seoul'],
'fruits': ['Apple', 'Orange', 'Banana',
'Banana', 'Apple', 'Apple', 'Banana'],
'price': [100, 200, 250, 300, 150, 200, 400],
'quantity': [1, 2, 3, 4, 5, 6, 7]
})
df
```

```python
x = df.groupby('city')['price'].mean()
x
```

```python
x = df.groupby(['city','fruits']).mean()
x
```

### map()

```python
df = pd.DataFrame({
'city': ['Busan', 'Busan', 'Busan', 'Busan',
'Seoul', 'Seoul', 'Seoul'],
'fruits': ['Apple', 'Orange', 'Banana',
'Banana', 'Apple', 'Apple', 'Banana'],
'price': [100, 200, 250, 300, 150, 200, 400],
'quantity': [1, 2, 3, 4, 5, 6, 7]
})
df
```

```python
z = df['city'].map(lambda x:'SEO' if x=='Seoul' else 'BUS')
df['TAG'] = z
df
```

```python
df['TAG'] = df['TAG'].map({'BUS':0, 'SEO':1})
df
```
