# Array(배열)
  - 다차원 배열
  - 배열의 속성
  - 배열의 자료형 (Data Type)
  - 배열 함수와 생성함수
  - Shape 변환 함수
  - 배열의 형태 (Shape)와 축(Axis) 지정
  - 인덱싱 (indexing), 슬라이싱 (Slicing)
  - 산술연산과 Broadcasting(브로드캐스팅)

```python
d1 = [1,2,'3',4,5]

print(type(d1), type(d1[1]), type(d1[2]))
```

```python
a1 = np.array(d1)
print(a1)
print(type(a1),a1.shape, a1.dtype, a1.itemsize)
#shape 반환타입 - 튜플
```

- 형변환처리

```python
d8 = list(range(1,13))
d8
a8_1 = np.array(d8)
print(a8_1)
a8_2 = np.array(d8).reshape(2,6)
print(a8_2)
a8_3 = np.array(d8).reshape(3,2,2)
print(a8_3)

print(a8_1.shape, a8_2.shape, a8_3.shape)
print(a8_1.ndim, a8_2.ndim, a8_3.ndim)

print(a8_3[0][1][0], a8_3[1][1][1],a8_3[2][1][0])
```

```python
a9 = np.arange(1,7).reshape(2,3)
print(a9,a9.dtype)
a9 = a9.astype(np.float64)
print(a9,a9.dtype)
a9 = a9.astype('int32')
print(a9, a9.dtype)
a9 = a9.astype('float64')
print(a9, a9.dtype)
a9 = a9.astype('str')
print(a9, a9.dtype, a9.itemsize)
a9 = a9.astype(np.bytes_) # np.string_ 타입 사용 불가 -> 대신 np.bytes_사용
print(a9,a9.dtype, a9.itemsize)

a9 = a9.astype(np.str_)
print(a9,a9.dtype, a9.itemsize)
```

- functions

```python
a1 = np.arange(1,26).reshape(5,5)
print(a1)

print(a1[2][2], a1[2,2])
print(a1[1:-1])
print(a1[1:-1,1:-1])
print(a1[3:,:3])
print(a1[:1,-1:])

a2 = np.ones(9)
print(a2)

a2 = np.ones(9).reshape(3,3)
print(a2)

a3 = np.zeros(9).reshape(3,3)
print(a3)

a3 = np.eye(5)
print(a3)
```

```python
a4 = np.arange(1,7).reshape(2,3)
print(a4)

a5 = np.transpose(a4)
print(a5)

a6 = a5.T
print(a6)
```

```python
a1 = np.array([[1,2], [3,4]])
a2 = np.array([10,20])
a3 = np.array([3,4])

print(a2+1)
print(a2-30)
print(a2*10)
print(a2/2)
print(100/a2)

print(a1+1)
print(a1-30)
print(a1*10)
print(a1/2)
print(100/a1)

print(a2+a3)
print(a1+a2)

print(a1*a2)
```

```python
a3 = np.array([[1,2],[3,4]])
a4 = np.array([[2,3],[4,5]])

print(a3+a4)
print(a3*a4)
print(np.matmul(a3,a4))

a5 = np.arange(1,101)
print(a5.max(), a5.min(), a5.mean(), a5.sum())

a5 = np.arange(1,101).reshape(10,10)
print(a5.max(), a5.min(), a5.mean(), a5.sum())
```
